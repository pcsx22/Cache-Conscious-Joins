```python
import os
import sys
import pyperf
import time
import matplotlib.pyplot as plt
```

```python
def getRunCommand(x):
    return '/home/sushil/Desktop/csc586/main ' + str(x) + ' file1 file2 =' 
def getRunCommandWithPartition(x, t):
    return '/home/sushil/Desktop/csc586/main ' + str(x) + ' file1 file2 = ' + str(t) 
```

```python
def getRunTime(inputSize):
    numIterations = 1
    commands = [getRunCommand(x) for x in inputSize]
    runTime = []
    for cmd in commands:
        avgTime = float(0)
        for i in range(numIterations):
            t = time.time()
            os.system(cmd)
            avgTime += time.time() - t
        runTime.append(avgTime/numIterations)
    return runTime
```

```python
#with new algo
inputSize = [100, 1000, 10000, 100000]
runTimePartitioned = getRunTime(inputSize)
```

```python
runTimeBasic = getRunTime(inputSize)
```

```python
plt.rcParams["figure.figsize"] = (18,5)
fig, ax = plt.subplots()
ax.plot(inputSize, runTimePartitioned, marker='x', label='Partitioned Sort')
ax.plot(inputSize, runTimeBasic, marker = 'o', label='Basic Sort')
ax.legend()
```
